```python
import requests
import pandas as pd
import numpy as np
import json
import matplotlib
from getpass import getpass
```

```python
url = "https://api.github.com"
header = {"Accept": "application/vnd.github.v3+json",
          "Authorization": f"token {getpass()}"}
response = requests.get(url, header)
response.status_code
```

```python
curr = url + "/repos/ironhack-datalabs/data-ber-08-20/forks"
response = requests.get(curr, header)
response.status_code
```

```python
forks = response.json()
forks
```

```python
l_urls = [v for fork in forks for k, v in fork.items() if k == "languages_url"]
langs = set()
for url in l_urls:
    response = requests.get(l_urls[0], header)
    langs.update({language for language in response.json().keys()})
langs
```
